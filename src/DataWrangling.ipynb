# Data Wrangling

```python
books = pd.read_json("./data/books.json")
metadata = pd.read_json('./data/metadata_nosuffix.json')
```

```python
books.head()
```

```python
metadata.head()
```

```python
metadata[metadata['index'] == 10].subjects_new
```

```python
classB = []
for i in range(len(metadata)):
    for sub in metadata.iloc[i].subjects_new:
        if sub in set(['B', 'BC', 'BD', 'BF', 'BH', 'BJ', 'BL', 'BM', 'BP', 'BQ', 'BR', 'BS', 'BT', 'BV', 'BX']):
            classB.append(metadata.iloc[i, 0])
            break
```

```python
books_B = books[books['index'].isin(classB)]
```

```python
books_B.to_json('./data/books_B.json')
```

```python
metadata_B = metadata[metadata['index'].isin(classB)]
```

```python
metadata_B.to_json('./data/metadata_B.json')
```
