# Data Wrangling

```python
dataset = pd.read_json('./data/fulldata.json')
```

```python
window = 4096
skip = 300
X, y = [], []
for i in range(len(dataset)):
    tokens = ' '.join(dataset.text[i].split()[skip:4096+skip])
    X.append(tokens)
    temp = []
    for s in dataset.subjects_new[i]:
        if s in set([
                        'B', 'BC', 'BD', 'BF', 'BH', 'BJ', 'BL', 'BM', 'BP', 'BQ', 'BR', 'BS', 'BT', 'BV', 'BX'
                    ]):
            temp.append(s)
    y.append(temp)
```

```python
multi = []
for i in range(len(y)):
    if len(y[i]) > 1:
        multi.append(i)
        if len(y[i]) > 2:
            print(len(y[i]))
```

```python
final_X, final_y, final_index = [], [], []
for i in range(len(y)):
    if y[i] != []:
        if i in multi:
            final_y.append(y[i][0])
            final_X.append(X[i])
            final_index.append(dataset.id[i])
            final_y.append(y[i][1])
            final_X.append(X[i])
            final_index.append(dataset.id[i])
            try:
                final_y.append(y[i][2])
                final_X.append(X[i])
                final_index.append(dataset.id[i])
            except IndexError:
                pass
        else:
            final_y.append(y[i][0])
            final_X.append(X[i])
            final_index.append(dataset.id[i])
```

```python
len(final_X), len(final_y), len(final_index)
```

```python
df = pd.DataFrame({'id': final_index, 'X': final_X, 'y': final_y})
df.head()
```

```python
df.to_json('./data/dataset_B.json')
```
