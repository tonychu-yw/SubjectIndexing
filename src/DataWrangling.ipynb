# Data Wrangling

```python
dataset_B = pd.read_json('./data/dataset_B.json')
```

```python
dataset_B.head()
```

```python
window = 4096
skip = 300
X = []
y = []
for i in range(len(dataset_B)):
    tokens = ' '.join(dataset_B.text[i].split()[skip:4096+skip])
    X.append(tokens)
    temp = []
    for s in dataset_B.subjects_new[i]:
        if s in set(['B', 'BC', 'BD', 'BF', 'BH', 'BJ', 'BL', 'BM', 'BP', 'BQ', 'BR', 'BS', 'BT', 'BV', 'BX']):
            temp.append(s)
    y.append(temp)
```

```python
len(X)
```

```python
multi = []
for i in range(len(y)):
    if len(y[i]) > 1:
        multi.append(i)
        if len(y[i]) > 2:
            print(len(y[i]))
```

```python
final_y = []
for i in range(len(y)):
    if i in multi:
        final_y.append(y[i][0])
        final_y.append(y[i][1])
    else:
        final_y.append(y[i][0])

final_X = []
for i in range(len(X)):
    if i in multi:
        final_X.append(X[i])
        final_X.append(X[i])
    else:
        final_X.append(X[i]) 
```

```python
len(final_X), len(final_y)
```

```python
df = pd.DataFrame({'X': final_X, 'y': final_y})
```

```python
df.to_json('./data/dataset_B.json')
```
