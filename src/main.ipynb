```python
# count subject frequencies
counter = {}
for row in metadata.subjects_new:
    for sub in row:
        if sub not in counter:
            counter[sub] = 1
        else:
            counter[sub] += 1
counter_tuples = sorted(counter.items(), key=lambda x: x[1], reverse=True)
```
