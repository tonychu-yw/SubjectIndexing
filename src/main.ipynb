```python
# prepare metadata
with open("./data/books_B.json", "r") as f:
    books_B = json.load(f) 
books_B = {int(k):v for k,v in books_B.items()}  # convert str keys to int
books_B = dict(sorted(books_B.items()))  # sort books
metadata = pd.read_json('./data/metadata.json')
# metadata_B = metadata[metadata['document'].isin(list(books_B.keys()))]  # filter books in class B
# metadata_B = metadata_B.sort_values(by=['document'])  # sort metadata

# remove suffix
new_subjects = []
for row in metadata.subjects:
    clean_subject = []
    for subject in row:
        suf = subject.find(" --")
        if suf == -1:
            sub = subject
        else:
            sub = subject[:suf]
        if sub != 'B':
            clean_subject.append(sub)
    if clean_subject == []:
        clean_subject.append('')
    new_subjects.append(clean_subject)
```

```python
metadata['subjects_new'] = new_subjects
```
